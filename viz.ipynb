```python
files = glob.glob('velocity*.txt')
nums = np.zeros(len(files))
for i,f in enumerate(files):
    nums[i]=f.split('Nb')[1].split('_')[0]
files=np.array(files)[np.argsort(nums)]
```

```python
arr =np.loadtxt(files[0],delimiter =" ")
arrs=np.zeros((6,len(arr),2))
for i,f in enumerate(files):
    arrs[i] =np.loadtxt(f,delimiter =" ")
    plt.plot(arrs[i,:,1])
```

```python
plt.plot(arrs[0,:,1])
```

```python
from sklearn.linear_model import LinearRegression

fits = []
for a in range(len(arrs)):
    model = LinearRegression(fit_intercept=True)

    model.fit(arrs[a,:,0,np.newaxis],arrs[a,:,1])

    xfit = np.linspace(0, 10, 1000)
    yfit = model.predict(arrs[0,:,0,np.newaxis])
    fits.append(model.coef_[0])
    plt.plot(((yfit-arrs[a,:,1])**2)[20:])
```

```python
plt.plot([10,20,50, 100,200,500],fits)
```
